## Data Visualization

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
sns.set_style('darkgrid')
```

```python
font1 = {'family':'serif', 'color':'black', 'size':20}
font2 = {'family':'serif', 'color':'darkred', 'size':15}

sns.set_context('paper', font_scale=1)
```

### max_weather_conditions

```python
max_weather_cond.plot()
plt.rcParams['figure.figsize'] = (15, 10)
plt.title('Weather Conditions Max Values', fontdict=font1)
plt.grid(axis='both', color='black')
plt.xlabel('Weather Condition', fontdict=font2)
plt.show()
```

```python
weathers = [weathers for weathers, df in max_weather_cond.groupby('Weather Condition')]
y = max_weather_cond['Visibility_km']

plt.bar(weathers, y, width=0.7, align='center', label='Weather')
plt.title('Weather Conditions And Max Of Visibility Level', fontdict=font1)
plt.xticks(weathers, rotation=90)
plt.xlabel('Weather Condition', fontdict=font2)
plt.ylabel('Visibility Level(km)', fontdict=font2)
plt.grid(True, color='black')
plt.figure(figsize=(40,60))
plt.show()
```
