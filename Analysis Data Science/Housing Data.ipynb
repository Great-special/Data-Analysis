### Questions 

#### Q1. Convert the Datatype of 'Date' column to Date-Time format.

```python
housing_data.head(2)
```

```python
housing_data.date = pd.to_datetime(housing_data.date)
```

```python
housing_data.dtypes
```

```python
housing_data=housing_data.drop(columns=['year'])
```

#### Q2. (A) Add a new column 'Year' in the dataframe, which contains years only.

Using the dataframe.[column_name].dt.year

```python
housing_data['year'] = housing_data.date.dt.year

housing_data
```

#### Q2. (B) Add a new column 'month' as the 2nd column in the dataframe, which contains month only.

using the .insert(col_index, col_name, col_values) function to place the column where you want it 

```python
housing_data.head(2)
```

```python
housing_data.insert(1, 'month', housing_data.date.dt.month)
```

```python
housing_data
```

#### Q4. What is maximum & minimum 'average_price' per year in england?

```python
england_data = housing_data[housing_data.area == 'england']
```

```python
england_data
```

```python
england_data_max = england_data.groupby('year')[['average_price']].max()
england_data_max
```

```python
england_data_min = england_data.groupby('year')[['average_price']].min()
england_data_min
```

#### Q5. What is the Maximum & Minimum No Of Crimes recorded per area?

Sorting them in ascending order

```python
area_data_max = housing_data.groupby('area')[['no_of_crimes']].max()
area_data_max.sort_values(ascending=True, by='no_of_crimes')
```

```python
area_data_min = housing_data.groupby('area')[['no_of_crimes']].min()
area_data_min.sort_values(ascending=True, by='no_of_crimes')
```

#### Q6. Show the total count of records of each area, where average price is less than 100000

```python
housing_data[housing_data.average_price < 100000]['area'].value_counts()
```
